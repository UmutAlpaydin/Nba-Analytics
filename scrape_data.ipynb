```python
# Scrape per game data for 2021-2022 season by pandas
year = '2022'
str = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

url = str.format(year)
url
```

```python
import pandas as pd
```

```python
df = pd.read_html(url, header= 0)
df
```

```python
df[0]
```

```python
df2022 = df[0]
```

```python
df = df2022.drop(df2022[df2022.Age == 'Age'].index)
```
